# **1. Data Cleaning**

```python
# Import necessary libraries
import pandas as pd
import numpy as np

retail_sales = pd.read_csv('/content/retail_sales_dataset.csv')
menu = pd.read_csv('/content/menu.csv')

retail_sales_df = pd.DataFrame(retail_sales)
menu_df = pd.DataFrame(menu)

# Check for missing values
print("\nMissing Values in Retail Sales Dataset:")
print(retail_sales_df.isnull().sum())

print("\nMissing Values in Menu Dataset:")
print(menu_df.isnull().sum())
```

```python
# Handle missing values using forward fill
retail_sales_df.fillna(method='ffill', inplace=True)
menu_df.fillna(method='ffill', inplace=True)

# Check for duplicate rows
print("\nDuplicate rows in Retail Sales Dataset:", retail_sales_df.duplicated().sum())
print("Duplicate rows in Menu Dataset:", menu_df.duplicated().sum())
```

```python
# Remove duplicates
retail_sales_df.drop_duplicates(inplace=True)
menu_df.drop_duplicates(inplace=True)

# Display the cleaned datasets (head)
print("\nCleaned Retail Sales Dataset:")
print(retail_sales_df.head())
```

```python
print("\nCleaned Menu Dataset:")
print(menu_df.head())
```

# **2. Descriptive Statistics**

```python
# Retail Sales Dataset Descriptive Statistics
print("\nDescriptive Statistics for Retail Sales Dataset:")
print(retail_sales_df.describe(include='all'))
```

```python
# Menu Dataset Descriptive Statistics
print("\nDescriptive Statistics for Menu Dataset:")
print(menu_df.describe(include='all'))
```

# **3. Time Series Analysis**

```python
import matplotlib.pyplot as plt

# Convert Date column to datetime (if present)
if 'Date' in retail_sales_df.columns:
    retail_sales_df['Date'] = pd.to_datetime(retail_sales_df['Date'])

# Group sales by month and analyze trends
if 'Date' in retail_sales_df.columns and 'Total Amount' in retail_sales_df.columns:
    monthly_sales = retail_sales_df.groupby(retail_sales_df['Date'].dt.to_period('M'))['Total Amount'].sum()
    monthly_sales_df = monthly_sales.reset_index()
    print("\nMonthly Sales Trends:")
    print(monthly_sales_df)

    # Plot monthly sales trend
    plt.figure(figsize=(10, 6))
    monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend', color='blue')
    plt.ylabel('Sales')
    plt.xlabel('Month')
    plt.grid()
    plt.show()
```

# **4. Customer and Product Analysis**

```python
# Product-wise sales analysis
if 'Product Category' in retail_sales_df.columns and 'Total Amount' in retail_sales_df.columns:
    product_sales = retail_sales_df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
    product_sales_df = product_sales.reset_index()
    print("\nTop 5 Products by Sales:")
    print(product_sales_df.head(5))

    # Bar plot for top 10 products
    plt.figure(figsize=(12, 6))
    product_sales.head(10).plot(kind='bar', color='skyblue', title='Top 10 Products by Sales')
    plt.ylabel('Total Sales')
    plt.xlabel('Product')
    plt.grid(axis='y')
    plt.show()


```

```python
# Category-wise average calorie analysis
category_calories = menu_df.groupby('Category')['Calories'].mean()
category_calories_df = category_calories.reset_index()
print("\nAverage Calories by Category:")
print(category_calories_df)

# Bar plot for category average calories
plt.figure(figsize=(12, 6))
category_calories.plot(kind='bar', color='coral', title='Average Calories by Category')
plt.ylabel('Average Calories')
plt.xlabel('Category')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# **5. Visualization (Correlation Heatmap)**

```python
import seaborn as sns

# Correlation heatmap for numerical columns in Retail Sales Dataset
numeric_columns = retail_sales_df.select_dtypes(include=['number'])
if not numeric_columns.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
    plt.title('Correlation Heatmap (Retail Sales Dataset)')
    plt.show()
else:
    print("\nNo numeric columns available for correlation heatmap in the Retail Sales Dataset.")
```

# **6. Recommendations**

```python
def retail_sales_recommendations(retail_sales_df):
    recommendations = []

    # 1. Customer Segmentation
    if 'Age' in retail_sales_df.columns:
        age_segments = pd.cut(retail_sales_df['Age'], bins=[0, 18, 35, 50, 100], labels=['Teens', 'Young Adults', 'Middle Age', 'Seniors'])
        top_segment = age_segments.value_counts(normalize=True).idxmax()
        recommendations.append(f"Target {top_segment} segment for marketing")

    # 2. Gender Analysis
    if 'Gender' in retail_sales_df.columns:
        gender_spending = retail_sales_df.groupby('Gender')['Total Amount'].sum()
        top_gender = gender_spending.idxmax()
        recommendations.append(f"Focus marketing efforts on {top_gender} customers")

    # 3. Product Category Insights
    if 'Product Category' in retail_sales_df.columns:
        category_sales = retail_sales_df.groupby('Product Category')['Total Amount'].agg(['sum', 'count'])
        top_category = category_sales['sum'].idxmax()
        recommendations.append(f"Expand {top_category} product line")

    # 4. Quantity and Pricing
    if 'Quantity' in retail_sales_df.columns and 'Price per Unit' in retail_sales_df.columns:
        avg_quantity = retail_sales_df['Quantity'].mean()
        recommendations.append(f"Optimize inventory for average order quantity of {avg_quantity:.2f}")

    return recommendations

def menu_dataset_recommendations(menu_df):
    recommendations = []

    # 1. Nutritional Category Analysis
    category_calories = menu_df.groupby('Category')['Calories'].mean()
    lowest_calorie_category = category_calories.idxmin()
    highest_calorie_category = category_calories.idxmax()
    recommendations.append(f"Promote {lowest_calorie_category} as health-conscious option")
    recommendations.append(f"Review portion sizes for {highest_calorie_category}")

    # 2. Nutritional Balance
    if 'Protein' in menu_df.columns:
        high_protein_items = menu_df[menu_df['Protein'] > menu_df['Protein'].mean()]
        recommendations.append(f"Highlight {len(high_protein_items)} high-protein menu items")

    # 3. Dietary Considerations
    if 'Dietary Fiber' in menu_df.columns:
        fiber_analysis = menu_df.groupby('Category')['Dietary Fiber'].mean()
        top_fiber_category = fiber_analysis.idxmax()
        recommendations.append(f"Emphasize {top_fiber_category} for dietary fiber intake")

    # 4. Vitamin and Mineral Content
    vitamin_columns = ['Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                       'Calcium (% Daily Value)', 'Iron (% Daily Value)']
    high_vitamin_items = menu_df[menu_df[vitamin_columns].mean(axis=1) > 50]
    recommendations.append(f"Promote {len(high_vitamin_items)} nutrient-rich menu items")

    return recommendations

# LETS START THE RECOMMENDATIONS
retail_recommendations = retail_sales_recommendations(retail_sales_df)
menu_recommendations = menu_dataset_recommendations(menu_df)

print("Retail Sales Recommendations:")
for rec in retail_recommendations:
    print(rec)

print("\nMenu Dataset Recommendations:")
for rec in menu_recommendations:
    print(rec)
```
